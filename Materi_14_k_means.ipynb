```python
#K Means with Python
import numpy as np
import matplotlib.pyplot as plt


X = -2 * np.random.rand(100,5)
X1 = 1+2 * np.random.rand(50,5)
X[50:100,:] = X1

plt.scatter(X[:,0],X[:,1],s=50,c='b')
plt.show()
```

```python
#STEP 1 : menentukan elbow yakni sharp turn

from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init = 'k-means++',max_iter=300, n_init=10, random_state = 0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Step 2 menetukan epicenter
Kmean = KMeans(n_clusters=2) #2 = elbow
Kmean.fit(X)

KMeans(algorithm='auto',
       copy_x=True,
       init='k-means++',
       max_iter=300,
       n_clusters=2,
       n_init=10,
       random_state=None, 
       tol=0.0001,
       verbose=0)

Kmean.cluster_centers_

```

```python
plt.scatter(X[:,0],X[:,1],s=50,c='b')
plt.scatter(Kmean.cluster_centers_[0,0],Kmean.cluster_centers_[0,1],s=200,c='g',marker='s')
plt.scatter(Kmean.cluster_centers_[1,0],Kmean.cluster_centers_[1,1],s=200,c='r',marker='s')
plt.show()
```
